# Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
```

# Basic Graph 

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8,]

#plt.plot(x,y, label='3x', color = 'green', linewidth=2, marker="x", markersize=10, markeredgecolor='yellow')

# Resize your graph
#figsize(width, height)
#dpi is pixels per inch
plt.figure(figsize=(5,3), dpi= 100)

# Use shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'gx-', label='2x')

#Line 2
#select interval to plot points
x2 = np.arange(0,2.5,0.5)

# plot part of graph as line
plt.plot(x2[:4], x2[:4]**2, 'y', label='x^2')

# plot remainder of graph as dot 
plt.plot(x2[3:], x2[3:]**2, 'b--')

# add title to graph
plt.title('Our First Graph', fontdict={'fontname': 'Arial', 'fontsize': 20})

# x and y label
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4,5])
#plt.yticks([0,2,4,6,8,10])

plt.legend()

#save file in here
plt.savefig('mygraph.png', dpi=300)

#show plot
plt.show()
```

# Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels,values)

patterns = ['x', '*', 'O']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
    
#bars[0].set_hatch('/')
#bars[1].set_hatch('x')
#bars[2].set_hatch('O')

plt.figure(figsize=(6,4))

plt.show()
```

# Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

# Histograms

```python
import matplotlib.pyplot as plt

bins = [50,60,70,80,90,100,110]

plt.hist(fifa.Overall, bins=bins, color='y')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA')



plt.show()
```

# Pie Chart

```python
real = fifa.loc[fifa['Real Face'] == 'Yes'].count()[1]

fake = fifa.loc[fifa['Real Face'] == 'No'].count()[1]

labels = ['Yes', 'No']
colors = ['blue', 'red']

plt.title('Face Validities of Players')

plt.pie([real, fake], labels = labels, colors = colors, autopct='%.2f %%')

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

lightweight = fifa.loc[fifa.Weight < 125].count()[0]
super_lightweight = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
welterweight = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
super_welterweight = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
middleweight = fifa[(fifa.Weight >= 200)].count()[0]

plt.title('Weight Distribution of FIFA Players (in lbs)')

weights = [lightweight, super_lightweight, welterweight, super_welterweight, middleweight]
labels = ['Below 125', '125 - 150', '150 - 175', '175 - 200', 'Above 200']
explode = (.1, .3 , 0.2, 0.2, 0.1)

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.5, explode=explode)

plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(7,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
arsenal = fifa.loc[fifa.Club == 'Arsenal']['Overall']


labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution', 'Arsenal']



boxes = plt.boxplot([barcelona, madrid, revs, arsenal], labels=labels, patch_artist=True, medianprops={'linewidth':2})


for box in boxes['boxes']:
    # set edge color
    box.set(color= 'green', linewidth = 2)

    # Change fill color
    box.set(facecolor='blue')
     
plt.title('Professional Football Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```
